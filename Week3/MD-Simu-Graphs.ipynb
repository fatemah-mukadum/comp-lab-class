```python
# install gromacs library
%pip install gromacs
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load potential xvg file 
time, potential = np.loadtxt('Data/potential.xvg', comments=["@", "#"],unpack=True)
time2, temperature = np.loadtxt('Data/temperature.xvg', comments=["@", "#"],unpack=True)
```

```python
# plot time vs potential 
# add labels and title
plt.plot(time, potential)
plt.xlabel('Energy Minimization Step')
plt.ylabel('Potential Energy (kJ/mol)')
plt.title('Energy Minimization')
#make plot look nice
plt.tight_layout()
# use scientific notation for y axis
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
# use arial font and make bigger
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 12
```

```python
# plot time2 vs temperature
# add labels and title
plt.plot(time2, temperature)
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Temperature')
#make plot look nice
plt.tight_layout()
```
