```python
# install gromacs library
%pip install gromacs
import numpy as np
import matplotlib.pyplot as plt
```

```python
time, PE, KE, totalE, Temp, Pressure, Volume = np.loadtxt("../Data/1hz3_T310.run.25000000.energy.xvg",comments=["@", "#"],unpack=True)
# convert time from ps to ns
time = time/1000
```

```python
def plotVar(x, x_label, y, y_label, name, a): 
    a.set(xlabel = x_label, ylabel = y_label)
    a.set_title(name)
    a.plot(x,y)
```

```python
# plot each quantity vs time on different plots
fig, axs = plt.subplots(6, figsize=(8, 40))
plotVar(time, "Time (ns)", PE, "Potential Energy (kJ/mol)", "Time vs Potential Energy", axs[0])
plotVar(time, "Time (ns)", KE, "Kinetic Energy (kJ/mol)", "Time vs Kinetic Energy", axs[1] )
plotVar(time, "Time (ns)", totalE, "Total Energy (kJ/mol)", "Time vs Total Energy" , axs[2])
plotVar(time, "Time (ns)", Temp, "Temperature (K)", "Time vs Temperature" , axs[3])
plotVar(time, "Time (ns)", Pressure, "Pressure (bar)", "Time vs Pressure" , axs[4])
plotVar(time, "Time (ns)", Volume, "Volume (nm^3)", "Time vs Volume", axs[5] )
plt.savefig("QuantVSTime.pdf", format="pdf", bbox_inches="tight")
```

```python
fig, ax1 = plt.subplots(figsize=(8, 8))
ax2 = ax1.twinx()
ax1.plot(time, PE, color='k', label="Potential Energy")
ax2.plot(time, KE, color='g', label="Kinetic Energy")
ax1.set_xlabel("Time (ns)")
ax1.set_ylabel("Potential Energy (kJ/mol)")
ax2.set_ylabel("Kinetic Energy (kJ/mol)")
ax1.legend(loc=2)
ax2.legend(loc=1)
plt.savefig("EnergyVSTime.pdf", format="pdf", bbox_inches="tight")
```
