```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time, PE, KE, totalE, Temp, Pressure, Volume = np.loadtxt("../Data/1hz3_T310.run.25000000.energy.xvg",comments=["@", "#"],unpack=True)
```

```python
def plotVar(x, x_label, y, y_label, name): 
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(name)
    plt.plot(x,y)
```

```python
plotVar(time, "Time (ps)", PE, "Potential Energy (kJ/mol)", "Time vs Potential Energy" )
```

```python
plotVar(time, "Time (ps)", PE, "Kinetic Energy (kJ/mol)", "Time vs Kinetic Energy" )
```

```python
fig, ax1 = plt.subplots(figsize=(8, 8))
ax2 = ax1.twinx()

ax1.plot(time, PE, color='k')
ax2.plot(time, KE, color='g')
```
