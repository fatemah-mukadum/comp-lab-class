```python
# install gromacs library and import it
%pip install gromacs
import gromacs
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load mdtraj
import mdtraj as md
# load gro file
traj = md.load('Inputs/ala300.trr', top='Inputs/ala300.gro')
```

```python
# compute phi and psi dihedral angle from trajectory
#phi_psi = md.compute_dihedrals(traj, np.array([[5, 6, 8, 14]]))
# compute psi angle from trajectory
psi = md.compute_psi(traj)
# compute phi angle from trajectory
phi = md.compute_phi(traj)
print(phi)
# compute phi and psi dihedral angle from trajectory
#phi_psi = md.compute_dihedrals(traj, np.array([[5, 6, 8, 14], [6, 8, 14, 15]]))
```

```python
# plot phi vs time 
plt.plot(traj.time, phi[1])
plt.xlabel('Time (ps)')
plt.ylabel('Phi (rad)')
plt.title('Phi vs Time')
plt.show()
```

```python
# plot psi vs time
plt.plot(traj.time, psi[1])
plt.xlabel('Time (ps)')
plt.ylabel('Psi (rad)')
plt.title('Psi vs Time')
plt.show()
```

```python
# compute a 2d histogram of phi and psi angles
H, xedges, yedges = np.histogram2d(phi[1].flatten(), psi[1].flatten(), bins=50)
# plot the -log of the histogram
plt.imshow(-np.log(H.T), extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], aspect='auto', origin='lower')
# add labeled colorbar
plt.colorbar(label='-log(Probability)')
```
