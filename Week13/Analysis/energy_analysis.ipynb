```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# plot potential energy vs time from outputs 
npt = np.loadtxt('../Outputs/NPT-1.ener', skiprows=1)
nvt = np.loadtxt('../Outputs/NVT-1.ener', skiprows=1)
```

```python
# plot potential energy vs time from outputs
plt.plot(npt[:,1], npt[:,4], label='NPT')
plt.plot(nvt[:,1], nvt[:,4], label='NVT')
# add labels 
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u)')
plt.title('Potential Energy vs Time')
plt.legend()
plt.savefig('nvtnpt.png', dpi=400)
```

```python
qm_mm = np.loadtxt('../Outputs/monitor/MONITOR-1.ener', skiprows=1)
plt.plot(qm_mm[:,1], qm_mm[:,4], label='QM/MM')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (a.u)')
plt.title('Potential Energy vs Time')
plt.legend()
plt.savefig('qm_mm.png', dpi=400)
```
