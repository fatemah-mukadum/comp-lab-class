```python
# this notebook is for plotting
# the end to end analysis
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
```

```python
# load xvg file 
time, o_msd = np.loadtxt('/home/fm1484/comp-class/comp-lab-class/Week4/Setup/waterbox/gromacs/OxygenMSD.xvg', comments=["@", "#"],unpack=True)
time, cl_msd = np.loadtxt('/home/fm1484/comp-class/comp-lab-class/Week4/Setup/waterbox/gromacs/ChlorineMSD.xvg', comments=["@", "#"],unpack=True)
time, Na_msd = np.loadtxt('/home/fm1484/comp-class/comp-lab-class/Week4/Setup/waterbox/gromacs/msd_Na.xvg', comments=["@", "#"],unpack=True)



```

```python
# plot msd vs time
plt.plot(time, o_msd)
plt.plot(time, cl_msd)
plt.plot(time, Na_msd)
plt.xlabel('Time (ps)')
plt.ylabel('MSD (nm^2)')
plt.title('Mean Square Displacement')
plt.legend(['Water Oxygen', 'Chlorine', 'Sodium'])
plt.show()

# save figure
plt.savefig("MSD.png", format="png", bbox_inches="tight")
```
