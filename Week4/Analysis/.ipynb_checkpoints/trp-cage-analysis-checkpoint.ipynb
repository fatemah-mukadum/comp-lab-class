```python
# this notebook is for plotting
# the end to end analysis
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
```

```python
# load beta amyloid structure
traj = md.load_xtc('/home/fm1484/comp-class/comp-lab-class/Week4/Setup/2JOF-AMBER/gromacs/md_concat_nowater.xtc', 
top='/home/fm1484/comp-class/comp-lab-class/Week4/Setup/2JOF-AMBER/gromacs/md_center-concat.gro')
```

```python
# compute the distance between the first and last atom
def end_end(traj):
    # get the first and last atom
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atom
    dist = np.sqrt(np.sum((first-last)**2, axis=1))
    return dist

# compute the distance between the first and last atom
dist = end_end(traj)

# compute the radius of gyration at each time step
rg = md.compute_rg(traj)

```

```python
# plot the end-end distance and radius of gyration vs time on the same plot
plt.plot(traj.time, dist, label='end-end distance')
plt.plot(traj.time, rg, label='radius of gyration')
plt.xlabel('time (ps)')
plt.ylabel('distance (nm)')
plt.legend()
plt.savefig("End-End-and-RGyr-TrpCage.png", format="png", bbox_inches="tight")
```
