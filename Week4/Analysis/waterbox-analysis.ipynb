```python
# this notebook is for plotting
# the end to end analysis
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
```

```python
# load xvg file 
time, o_msd = np.loadtxt('/home/fm1484/comp-class/comp-lab-class/Week4/Setup/waterbox/gromacs/OxygenMSD.xvg', comments=["@", "#"],unpack=True)
time, cl_msd = np.loadtxt('/home/fm1484/comp-class/comp-lab-class/Week4/Setup/waterbox/gromacs/ChlorineMSD.xvg', comments=["@", "#"],unpack=True)
```
